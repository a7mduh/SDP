```python
import pandas as pd
import sqlite3 as lite

```

```python
db_file_path = r"C:\Users\rekog\OneDrive\Desktop\SDP\prutor-deepfix-09-12-2017.db"

# Create a connection to the SQLite database
conn = lite.connect(db_file_path)
```

```python
query = "select code, errorcount from Code"
```

```python
df = pd.read_sql_query(query, conn)
columns = ['code', 'safety']
df.columns = ['code', 'safety']
df
```

```python
df.loc[df['safety'] > 0, 'safety'] = 'vulnerable'
df.loc[df['safety'] == 0, 'safety'] = 'safe'
```

```python
df['language'] = ['C'] * df.shape[0]
order = ['code', 'language', 'safety']
df = df[order]
df
```

```python
df.to_csv('prutor-deepfix.csv', index= False)
```
